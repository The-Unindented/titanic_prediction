```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
df=pd.read_csv("/content/titanic (2).csv")
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python

df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.drop('Name',axis=1,inplace=True)
df.drop('Ticket',axis=1,inplace=True)

df.drop_duplicates(inplace=True)
```

```python
df['Age']=df['Age'].astype(int)
```

```python
df.columns = df.columns.str.strip().str.lower()
df.head()
```

```python
df["survived"]=df["survived"].astype('category') # Changed 'Survived' to 'survived'
df["pclass"]=df["pclass"].astype('category')   # Changed 'Pclass' to 'pclass'
df["sibsp"]=df["sibsp"].astype('category')     # Changed 'SibSp' to 'sibsp'
```

```python
print("Missing values after cleaning")
df.isnull().sum()
print("Clean dataset")
df.head()
```

```python
#Data preprocessing
#DATA TRANSFORMATION
#One-hot encoding
df = pd.get_dummies(df, columns=['pclass'],prefix='class')
df.drop(columns=['Passengerid'],axis=1,inplace=True)
df.head()
df["sibsp"].dtype
#df["parch"].dtype
```

```python
#feature engineering
df['sibsp'] = df['sibsp'].astype(int)
df['parch'] = df['parch'].astype(int)
df['familysize'] = df['sibsp'] + df['parch']
```

```python
df['embarked']=df['embarked'].map({'S':0,'C':1,'Q':2}).astype(int)
df.head()
```

```python
df['sex']=df['sex'].map({'male':0,'female':1}).astype(int)
```

```python
df.to_csv("clean data", index=False)
df.head()
```

```python
y=df.drop("survived",axis=1)
```

```python
y=df["survived"]
df.drop(["passengerid"],axis=1, inplace=True)
```

```python
df.drop("survived",axis=1,inplace=True)
```

```python
df
```

```python
x_train , x_test,y_train, y_test = train_test_split(df,y,test_size=.2,random_state=42)
```

```python
model=RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
